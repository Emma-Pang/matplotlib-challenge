```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "resources/mouse_drug_data.csv"
clinical_trial_data_to_load = "resources/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
drug_data= pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset
merge_df=pd.merge(drug_data, mouse_data,  on='Mouse ID' )

# Display the data table for preview
merge_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_vol=merge_df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean()
mean_tumor_vol_df=pd.DataFrame(mean_tumor_vol)
#reset index
mean_tumor_vol_df.reset_index()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
#find standard error by groupby
sems=merge_df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem()
#convert to df
sems_df=pd.DataFrame(sems)
#reset index
sems_df=sems_df.reset_index()
sems_df
```

```python
# Minor Data Munging to Re-Format the Data Frames (means)
mean_drug_timepoint_df=pd.pivot_table(mean_tumor_vol_df, values='Tumor Volume (mm3)', index=['Timepoint'],
               columns=['Drug'], fill_value=0)
mean_drug_timepoint_df
```

```python
# Minor Data Munging to Re-Format the Data Frames (sems)
sem_drug_timepoint_df=pd.pivot_table(sems_df, values='Tumor Volume (mm3)', index=['Timepoint'],
               columns=['Drug'], fill_value=0)
sem_drug_timepoint_df
```

```python
xval=sems_df['Timepoint'].unique()
xval
```

```python
# Generate the Plot (with Error Bars)
#x axis
x_axis = xval
#plot data
fig, ax = plt.subplots(figsize=(8,5))
mean_drug_timepoint_df['Capomulin'].plot(c='r',linestyle=':',marker='o')
mean_drug_timepoint_df['Infubinol'].plot(c='b',linestyle=':',marker='^')
mean_drug_timepoint_df['Ketapril'].plot(c='g',linestyle=':',marker='s')
mean_drug_timepoint_df['Placebo'].plot(c='black',linestyle=':',marker='d')
#add error bars
ax.errorbar(x_axis, mean_drug_timepoint_df['Capomulin'],yerr=sem_drug_timepoint_df['Capomulin'], ls='none', color='r')
ax.errorbar(x_axis, mean_drug_timepoint_df['Infubinol'],yerr=sem_drug_timepoint_df['Infubinol'], ls='none', color='b')
ax.errorbar(x_axis, mean_drug_timepoint_df['Ketapril'],yerr=sem_drug_timepoint_df['Ketapril'], ls='none', color='g')
ax.errorbar(x_axis, mean_drug_timepoint_df['Placebo'],yerr=sem_drug_timepoint_df['Placebo'] , ls='none', color='black')
#make it nice
ax.set_ylabel("Tumor Volume (mm3)")
ax.legend()
ax.grid(axis='y')
plt.title("Tumor Response to Treatment")
# Save the Figure
plt.savefig('tumor_response_to_treatment.png')
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_data=merge_df.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean()
# Convert to DataFrame
mean_met_df=pd.DataFrame(mean_met_data)
mean_met_df
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_sems=merge_df.groupby(['Drug','Timepoint'])['Metastatic Sites'].sem()
# Convert to DataFrame
met_sems_df=pd.DataFrame(met_sems)
# Preview DataFrame
met_sems_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_met_timepoint_df=pd.pivot_table(mean_met_df, values='Metastatic Sites', index=['Timepoint'],
               columns=['Drug'], fill_value=0)
# Preview that Reformatting worked
mean_met_timepoint_df
```

```python
# Minor Data Munging to Re-Format the Data Frames (sems)
sem_met_timepoint_df=pd.pivot_table(met_sems_df, values='Metastatic Sites', index=['Timepoint'],
               columns=['Drug'], fill_value=0)
sem_met_timepoint_df
```

```python
# Generate the Plot (with Error Bars)
#x axis
x_axis = xval
fig, ax = plt.subplots(figsize=(8,5))
#add data
mean_met_timepoint_df['Capomulin'].plot(c='r',linestyle=':',marker='o')
mean_met_timepoint_df['Infubinol'].plot(c='b',linestyle=':',marker='^')
mean_met_timepoint_df['Ketapril'].plot(c='g',linestyle=':',marker='s')
mean_met_timepoint_df['Placebo'].plot(c='black',linestyle=':',marker='d')
#add error bars
ax.errorbar(x_axis, mean_met_timepoint_df['Capomulin'],yerr=sem_met_timepoint_df['Capomulin'], ls='none', color='r')
ax.errorbar(x_axis, mean_met_timepoint_df['Infubinol'],yerr=sem_met_timepoint_df['Infubinol'], ls='none', color='b')
ax.errorbar(x_axis, mean_met_timepoint_df['Ketapril'],yerr=sem_met_timepoint_df['Ketapril'], ls='none', color='g')
ax.errorbar(x_axis, mean_met_timepoint_df['Placebo'],yerr=sem_met_timepoint_df['Placebo'] , ls='none', color='black')
#make it nice
ax.set_ylabel("Metastatic Sites")
ax.grid(axis='y')
ax.legend()
plt.title("Metastic Spread During Treatment")
# Save the Figure
plt.savefig('metastic_spread_during_treatment.png')
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_data=merge_df.groupby(['Drug','Timepoint'])['Mouse ID'].count()
# Convert to DataFrame
mouse_data_df=pd.DataFrame(mouse_data)
# Preview DataFrame
mouse_data_df.reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames 
mouse_data_reformat=pd.pivot_table(mouse_data_df, values='Mouse ID', index=['Timepoint'],
               columns=['Drug'], fill_value=0)
mouse_data_reformat
```

```python
#turn original dataframe into percentage of survival over time
percent_survival_df=(mouse_data_reformat/mouse_data_reformat.iloc[0])*100
```

```python
# Generate the Plot (Accounting for percentages)
#x axis
x_axis = xval
fig, ax = plt.subplots(figsize=(8,5))
#add data
percent_survival_df['Capomulin'].plot(c='r',linestyle=':',marker='o')
percent_survival_df['Infubinol'].plot(c='b',linestyle=':',marker='^')
percent_survival_df['Ketapril'].plot(c='g',linestyle=':',marker='s')
percent_survival_df['Placebo'].plot(c='black',linestyle=':',marker='d')
#make it nice
ax.set_ylabel("Survival Rate (%)")
ax.grid()
ax.legend()
plt.title("Survival During Treatment")
# Save the Figure
plt.savefig('survival_during_treatment.png')
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change=((mean_drug_timepoint_df.iloc[9]-mean_drug_timepoint_df.iloc[0])/(mean_drug_timepoint_df.iloc[0]))*100
#turn into dataframe and reformat
percent_change_df=pd.DataFrame(percent_change)
percent_change_df=percent_change_df.reset_index()
percent_change_df=percent_change_df.rename(columns={0: "Percent Change"})
percent_change_df
```

```python
#create new dataframe to hold only necessary info for chart
percent_change_df2 = percent_change_df[percent_change_df["Drug"].isin(["Capomulin","Infubinol","Ketapril","Placebo"])]
#create barchart
barchart=plt.bar(percent_change_df2["Drug"],percent_change_df2["Percent Change"])
#make it pretty
barchart[0].set_color('r')
barchart[1].set_color('g')
barchart[2].set_color('g')
barchart[3].set_color('g')
plt.ylabel("(%) Tumor Volume Change")
plt.title("Tumor Change Over 45 Day Treatment")
plt.grid()
# Save the Figure
plt.savefig('tumor_change_over_treatment.png')
```

## Observations

 - Based on the findings in each graph it looks as if Capomulin and Ramicane are the only one drugs given to the mice that have had tumor sizes shrink over time. The metastic spread for both of those drugs is also notably the lowest.
 - The mouse survival rates of Capomulin and Ramicane are the highest.
 - Lastly based on the summary bar graph information, both Capomulin and Ramican are the only drugs to have a negative change in tumor volume over time.
